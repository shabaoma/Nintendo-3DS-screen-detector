```python
import numpy as np
from keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input
"~/venv/venv/lib/python3.6/site-packages/keras/preprocessing/image.py"
"~/venv/venv/lib/python3.6/site-packages/PIL/Image.py"
"~/venv/venv/lib/python3.6/site-packages/keras/applications/imagenet_utils.py"
```

```python
img = image.load_img('../data/raw_image/2019-02-05 22.55.46.jpg', target_size=(256, 256))
img
```

```python
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
p_array = preprocess_input(img_array.copy())
```

```python
r_array = img_array.copy()[..., ::-1] # 'RGB'->'BGR'
mean = [103.939, 116.779, 123.68]
r_array[..., 0] -= mean[0]
r_array[..., 1] -= mean[1]
r_array[..., 2] -= mean[2]
```

```python
print(img_array.shape, p_array.shape, r_array.shape)
for i in range(10):
    print(i, img_array[0][0][i], p_array[0][0][i], r_array[0][0][i])
```

```python
image.array_to_img(np.squeeze(r_array, axis=0))
```
